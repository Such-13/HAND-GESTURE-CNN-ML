```python
# Import TensorFlow and Keras
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout

# Image handling libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd

# Sklearn libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
import os
imagepaths = []
for dirname, _, filenames in os.walk("C:\\Users\\sucha\\Downloads\\PRODIGY_ML_04"):
    for filename in filenames:
        path = os.path.join(dirname, filename)
        if path.endswith("png"):
            imagepaths.append(path)
print(len(imagepaths))
```

```python
def img_plot(img_path):
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    print("Shape of the image is ", img_rgb.shape)
    plt.grid(False)
    plt.imshow(img_rgb, cmap='gray')
    plt.xlabel("Width")
    plt.ylabel("Height")
    plt.title("Image " + img_path)

img_plot(imagepaths[8000])
```

```python
X = []
y = []
for path in imagepaths[:19999]:
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img, (128, 128))
    X.append(img)
    
    category = path.split("\\")[-2]
    try:
        label = int(category.split("_")[0][1])
    except (IndexError, ValueError) as e:
        print(f"Error processing path {path}: {e}")
        continue
    y.append(label)

X = np.array(X)
X = X.reshape(len(X), 128, 128, 1)
y = np.array(y)
print("Images loaded: ", len(X))
print("Labels loaded: ", len(y))
print(y[0], imagepaths[0])
```

```python
from sklearn.model_selection import train_test_split
threshold = 0.3
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=threshold, random_state=42)
```

```python
model = Sequential()
model.add(Conv2D(32, (5,5), activation='relu', input_shape=(128,128,1)))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dropout(0.3))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5, batch_size=64, verbose=2, validation_data=(X_test, y_test))
model.save('handgesturerecog_model.h5')
```

```python
tLoss, tAccuracy = model.evaluate(X_test, y_test)
print('Test accuracy: {:2.2f}%'.format(tAccuracy*100))
prediction = model.predict(X_test)
np.argmax(prediction[0]), y_test[0]
```

```python
def validate_gestures(predictions_array, true_label_array, img_array):
    class_names = ["down", "palm", "l", "fist", "fist_moved", "thumb", "index", "ok", "palm_moved", "c"]
    plt.figure(figsize=(15,5))
    for i in range(1, 10):
        prediction = predictions_array[i]
        true_label = true_label_array[i]
        img = img_array[i]
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
        plt.subplot(3,3,i)
        plt.grid(False)
        plt.xticks([])
        plt.yticks([])
        plt.imshow(img, cmap=plt.cm.binary)
        predicted_label = np.argmax(prediction)
        color = 'blue' if predicted_label == true_label else 'red'
        plt.xlabel("Predicted: {} {:2.0f}% (Actual: {})".format(class_names[predicted_label],
                                   100*np.max(prediction),
                                   class_names[true_label]), color=color)
    plt.show()

validate_gestures(prediction, y_test, X_test)
y_pred = np.argmax(prediction, axis=1)
```

```python
def preprocess_image(img_path):
    img = cv2.imread(img_path)
    if img is None:
        print(f"Error: Unable to load image from path: {img_path}")
        return None
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img, (128, 128))
    img = img.reshape(1, 128, 128, 1)
    img = img.astype('float32') / 255.0
    return img

user_image_path = input("Enter the path to the image you want to predict: ")
processed_image = preprocess_image(user_image_path)
if processed_image is not None:
    single_prediction = model.predict(processed_image)
    predicted_class = np.argmax(single_prediction)
    class_names = ["down", "palm", "l", "fist", "fist_moved", "thumb", "index", "ok", "palm_moved", "c"]
    print(f"Predicted class: {class_names[predicted_class]} ({predicted_class})")

    def display_prediction(img_path, prediction):
        img = cv2.imread(img_path)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        plt.grid(False)
        plt.imshow(img_rgb, cmap='gray')
        plt.xlabel("Width")
        plt.ylabel("Height")
        plt.title(f"Predicted: {class_names[np.argmax(prediction)]} ({100*np.max(prediction):.2f}%)")
        plt.show()

    display_prediction(user_image_path, single_prediction)
```
